El botanista Robert Brown, observó al microscopio en 1827 un extraño movimiento que sufrían las partículas de polen que lo desconcertó, noto que las partículas se movían a través del líquido; pero no fue capaz de determinar los mecanismos que provocaron este movimiento. Sin embargo no fué sino hasta 1905 que Albert Einstein, estudiando este fenómeno, postuló que los granos de polen, a pesar de ser lo suficientemente grandes como para ser vistos en el microscopio, eran lo suficientemente pequeños como para que fuesen “golpeados” por la energía cinética-térmica de las moléculas que los rodeaban.

El movimiento browniano es un ejemplo de proceso estocástico simple, pero Robert Brown no fue quien explicó este fenómeno en un lenguaje matemático. Los fenómenos estocásticos empezaron a ser comprendidos a partir de los avances en la cinemática

En matemáticas, el movimiento browniano es descrito por el proceso de Wiener; un proceso estocástico de tiempo continuo nombrado en honor a Norbert Wiener. También conocido como el proceso de movimiento browniano, es un modelo matemático utilizado para describir el comportamiento aleatorio y continuo de las partículas pequeñas en un fluido.

Este se puede decir es un proceso estocástico continuo, lo que significa que su evolución a lo largo del tiempo es descrita por una función continua en lugar de una secuencia discreta de eventos. La trayectoria del proceso de Wiener se puede describir como una función que es continua pero no diferenciable en ningún punto.
